```python
from __future__ import division
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

os.chdir('/Users/mac/Desktop/Euroleageeks/Posts/Final4/Data')
Real = pd.read_csv('Real.txt')
CSKA = pd.read_csv('CSKA.txt')
Olympiacos = pd.read_csv('Olympiacos.txt')
Fenerbahce = pd.read_csv('Fenerbahce.txt')

teams = [Real, CSKA, Olympiacos, Fenerbahce]
names = ("ORTG", "DRTG", "NETRTG","POS", "eFG%", "DRB%", "ORB%" ,"FT Ratio", "TOV Ratio", "4Factor")
tn = ('Real', 'CSKA', 'Olympiacos', 'Fenerbahce')
t = pd.DataFrame(np.random.randn(4,10), columns=names, index = tn)


def var(t):
    global Team_MP
    global Team_DRB
    global Opponent_ORB
    global Team_BLK
    global Opponent_FGM
    global Opponent_FGA
    global Opponent_TOV
    global Team_STL
    global Opponent_FTA
    global Opponent_FTM
    global Team_PF
    global Team_Possessions
    global Opponent_MP
    global Opponent_PTS
    global Team_MP
    global Team_AST
    global Team_FGM
    global Team_PTS
    global Team_FTM
    global Team_FGA
    global Team_FTA
    global Team_TOV
    global Opponent_TRB
    global Opponent_ORB
    global Opponent_DRB
    global Team_ORB
    global THREEPM
    global Team_THREEPM
    Opponent_ORB = t.ix[1,15]
    Team_DRB=t.ix[0,16]
    Opponent_FGM=t.ix[1,3]
    Opponent_FGA=t.ix[1,4]
    Team_BLK=t.ix[0,20]
    Team_MP = t.ix[0, 2]
    Opponent_MP = t.ix[1, 2]
    Opponent_TOV = t.ix[1,21]
    Team_STL=t.ix[1,19]
    Team_PF=t.ix[0,22]
    Opponent_FTM=t.ix[1,12]
    Opponent_FTA=t.ix[1,13]
    Opponent_PTS = t.ix[1, 23]
    Team_MP=t.ix[0,2]
    Team_AST=t.ix[0,18]
    Team_FGM=t.ix[0,3]
    Team_PTS=t.ix[0,23]
    Team_FTM=t.ix[0,12]
    Team_FGA=t.ix[0,4]
    Team_FTA=t.ix[0,13]
    Team_ORB=t.ix[0,15]
    Team_TOV=t.ix[0,21]
    Team_THREEPM=t.ix[0,6]
    Opponent_TRB=t.ix[1,17]
    Opponent_ORB =t.ix[1,15]
    Opponent_DRB = t.ix[1, 16]

def Pos(t):
        var(t)
        return 0.5 * ((Team_FGA + 0.4 * Team_FTA - 1.07 * (Team_ORB / (Team_ORB + Opponent_DRB)) * (Team_FGA - Team_FGM) + Team_TOV) +
        (Opponent_FGA + 0.4 * Opponent_FTA - 1.07 * (Opponent_ORB / (Opponent_ORB + Team_DRB)) * (Opponent_FGA - Opponent_FGM) + Opponent_TOV))


for i in range(0,4):
    t.ix[i,3] = Pos(teams[i])/teams[i].ix[1,1]


def ORTG(x):
    var(x)
    return 100*(x.ix[0,23]/ (t.ix[i,3]*teams[i].ix[1,1]))


def DRTG(x):
    var(x)
    return 100 * (x.ix[1, 23] / (t.ix[i,3]*teams[i].ix[1,1]))


for i in range(0,4):
    t.ix[i,0]=ORTG(teams[i])
    t.ix[i, 1] = DRTG(teams[i])
    t.ix[i, 2] = t.ix[i,0] - t.ix[i, 1]

def efg(x):
    var(x)
    return (Team_FGM + 0.5 * Team_THREEPM)/ Team_FGA

def tov(x):
    var(x)
    return (Team_TOV) / (Team_FGA + 0.44 * Team_FTA + Team_TOV)

def ft(x):
    var(x)
    return (Team_FTM) / (Team_FGA)

def drb(x):
    var(x)
    return (Team_DRB) / (Opponent_ORB + Team_DRB)

def orb(x):
    var(x)
    return (Team_ORB) / (Opponent_DRB + Team_ORB)


for i in range(0,4):
    t.ix[i, 4] = efg(teams[i])
    t.ix[i, 5] = drb(teams[i])
    t.ix[i, 6] = orb(teams[i])
    t.ix[i, 7] = ft(teams[i])
    t.ix[i, 8] = tov(teams[i])
    t.ix[i, 9] = (0.4 * t.ix[i, 4]) - (0.25 * t.ix[i, 8]) + (0.1 * t.ix[i, 5])  + (0.1 * t.ix[i, 6]) + (0.15 * t.ix[i, 7])

n=np.arange(len(tn))

#barplot
barlist = plt.bar(n,t['4Factor'],align='center', width=0.6, alpha = 1)
plt.xticks(n,tn)
plt.ylabel('Weighted Score')
plt.title('Four factors of sucess rating')
barlist[0].set_color('#b942f4')
barlist[1].set_color('blue')
barlist[2].set_color('red')
barlist[3].set_color('#f1f441')

plt.ylim([0.28,0.35])
```
